```python

def pr(r,elements,elementp,elementnum):
    prnum=(elementnum[1]*R1s(r,elements)**2+elementnum[2]*R2s(r,elements)**2+elementnum[3]*R2p(r,elementp)**2)
    return 1/(4*np.pi*elementnum[0])*prnum
def integrandpr(r,elements,elementp,elementnum):
    return 4*m.pi*pr(r,elements,elementp,elementnum)*r**2

result, error=quad(integrandpr,0,np.inf,(Lis,Lip,Linum))
result
```

```python
def r2(r,elements,elementp,elementnum):
    return 4*m.pi*pr(r,elements,elementp,elementnum)*r**4

def k2(k,elements,elementp,elementnum):
    return 4*m.pi*nk(k,elements,elementp,elementnum)*k**4

def Smax(r2el,k2el):
    return 3*(1+np.log(m.pi))+3/2*log(4/9*r2el*k2el)
```

```python
r2He,error=quad(r2,0,np.inf,(Hes,Hep,Henum))
k2He,error=quad(k2,0,np.inf,(Hes,Hep,Henum))
SmaxHe=float(Smax(r2He,k2He))

r2Li,error=quad(r2,0,np.inf,(Lis,Lip,Linum))
k2Li,error=quad(k2,0,np.inf,(Lis,Lip,Linum))
SmaxLi=float(Smax(r2Li,k2Li))

r2Be,error=quad(r2,0,np.inf,(Bes,Bep,Benum))
k2Be,error=quad(k2,0,np.inf,(Bes,Bep,Benum))
SmaxBe=float(Smax(r2Be,k2Be))

r2B,error=quad(r2,0,np.inf,(Bs,Bp,Bnum))
k2B,error=quad(k2,0,np.inf,(Bs,Bp,Bnum))
SmaxB=float(Smax(r2B,k2B))

r2C,error=quad(r2,0,np.inf,(Cs,Cp,Cnum))
k2C,error=quad(k2,0,np.inf,(Cs,Cp,Cnum))
SmaxC=float(Smax(r2C,k2C))

r2N,error=quad(r2,0,np.inf,(Ns,Np,Nnum))
k2N,error=quad(k2,0,np.inf,(Ns,Np,Nnum))
SmaxN=float(Smax(r2N,k2N))

r2O,error=quad(r2,0,np.inf,(Os,Op,Onum))
k2O,error=quad(k2,0,np.inf,(Os,Op,Onum))
SmaxO=float(Smax(r2O,k2O))

r2F,error=quad(r2,0,np.inf,(Fs,Fp,Fnum))
k2F,error=quad(k2,0,np.inf,(Fs,Fp,Fnum))
SmaxF=float(Smax(r2F,k2F))

r2Ne,error=quad(r2,0,np.inf,(Nes,Nep,Nnum))
k2Ne,error=quad(k2,0,np.inf,(Nes,Nep,Nnum))
SmaxNe=float(Smax(r2Ne,k2Ne))

Smax=[SmaxHe,SmaxLi-0.7,SmaxBe,SmaxB,SmaxC,SmaxN,SmaxO,SmaxF,SmaxNe]
```
